```python
import pandas as pd
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
train_df = train_df.dropna()
train_df
```

```python
ids = test_df["Id"]
test_df = test_df.drop(columns = ["Id"])
train_df = train_df.drop(columns = ["Id"])
ids
```

```python
test_df
```

```python
train_df.groupby("C1").count()
```

```python
def change(df, column_name):
    values = df.groupby(column_name).count().index
    for x in range(len(values)):
        df[column_name] = df[column_name].replace(values[x], x)
    return df[column_name]
for x in train_df.dtypes.index:
    if train_df.dtypes[x] == "object":
        train_df[x] = change(train_df, x)
for x in test_df.dtypes.index:
    if test_df.dtypes[x] == "object":
        test_df[x] = change(test_df, x)
train_df
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
columns1 = train_df.drop('Target', axis = 1).columns
columns2 = test_df.columns
scaler = StandardScaler()
scaler.fit(train_df.drop('Target', axis = 1))
ndf = pd.DataFrame(scaler.transform(train_df.drop('Target', axis = 1)))
test_df = pd.DataFrame(scaler.transform(test_df))
scaler = MinMaxScaler()
scaler.fit(ndf)
ndf = pd.DataFrame(scaler.transform(ndf))
test_df = pd.DataFrame(scaler.transform(test_df))
for x in range(10):
    ndf = ndf.rename(columns = {x : columns1[x]})
    test_df = test_df.rename(columns = {x : columns2[x]})
ndf
```

```python
(train_df.groupby(["C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]).count() > 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
clf = RandomForestRegressor()
clf.fit(ndf, train_df['Target'])
pred = np.round(clf.predict(test_df), -2)
```

```python
pred
```

```python
newdf = pd.DataFrame()
newdf["Id"] = ids
newdf["Target"] = pred
newdf
```

```python
newdf.to_csv("sampleSubmission.csv", index = False)
```
