```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df
```

```python
tdf = pd.read_csv("test.csv")
tdf
```

```python
df.info()
```

```python
df = df.dropna()
df
```

```python
tdf.info()
```

```python
def change(df, column_name):
    values = df.groupby(column_name).count().index
    for x in range(len(values)):
        df[column_name] = df[column_name].replace(values[x], x + 1)
    return df[column_name]
for x in df.dtypes.index:
    if df.dtypes[x] == "object":
        df[x] = change(df, x)
df
```

```python
for x in tdf.dtypes.index:
    if tdf.dtypes[x] == "object":
        tdf[x] = change(tdf, x)
tdf
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

```python
plt.plot(list(df.sort_values("Target").Target))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('Target', axis = 1))
ndf = pd.DataFrame(scaler.transform(df.drop('Target', axis = 1)))
for x in range(10):
    columns = df.drop('Target', axis = 1).columns
    ndf = ndf.rename(columns = {x : columns[x]})
ndf
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(ndf)
ndf = pd.DataFrame(scaler.transform(ndf))
for x in range(10):
    columns = df.drop('Target', axis = 1).columns
    ndf = ndf.rename(columns = {x : columns[x]})
ndf
```

```python
scaler = StandardScaler()
scaler.fit(tdf)
ntdf = pd.DataFrame(scaler.transform(tdf))
scaler = MinMaxScaler()
scaler.fit(ntdf)
ntdf = pd.DataFrame(scaler.transform(ntdf))
for x in range(10):
    columns = tdf.columns
    ntdf = ntdf.rename(columns = {x : columns[x]})
ntdf
```

```python
from sklearn.model_selection import train_test_split
y = df['Target']
X_train, X_test, y_train, y_test = train_test_split(ndf, y)
X_train
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
clf = linear_model.LinearRegression()
clf.fit(ndf, df['Target'])
pred = clf.predict(X_test)
print(mean_squared_error(pred, y_test))
pred = clf.predict(X_train)
print(mean_squared_error(pred, y_train))
pred = clf.predict(ndf)
print(mean_squared_error(pred, df['Target']))
```

```python
clf = linear_model.Lasso(random_state=0, normalize = False)
clf.fit(ndf, df['Target'])
pred = clf.predict(X_test)
print(mean_squared_error(pred, y_test))
pred = clf.predict(X_train)
print(mean_squared_error(pred, y_train))
pred = clf.predict(ndf)
print(mean_squared_error(pred, df['Target']))
```

```python
from sklearn.model_selection import GridSearchCV
a = 10**np.linspace(10, -2, 100) * 0.5
p = {"alpha": a}
r = linear_model.Lasso()
rg = GridSearchCV(r, p, cv=5, n_jobs=-1, verbose=True)
rg.fit(df.drop('Target', axis = 1), df['Target'])
asd = rg.best_params_["alpha"]
rg.best_params_["alpha"]
```

```python
clf = linear_model.Lasso(random_state=0, alpha = rg.best_params_["alpha"])
clf.fit(ndf, df['Target'])
pred = clf.predict(X_test)
print(mean_squared_error(pred, y_test))
pred = clf.predict(X_train)
print(mean_squared_error(pred, y_train))
pred = clf.predict(ndf)
print(mean_squared_error(pred, df['Target']))
```

```python
clf = linear_model.Ridge(random_state=0)
clf.fit(ndf, df['Target'])
pred = clf.predict(X_test)
print(mean_squared_error(pred, y_test))
pred = clf.predict(X_train)
print(mean_squared_error(pred, y_train))
pred = clf.predict(ndf)
print(mean_squared_error(pred, df['Target']))
```

```python
from sklearn.model_selection import GridSearchCV
a = 10**np.linspace(10, -2, 100) * 0.5
p = {"alpha": a}
r = linear_model.Ridge()
rg = GridSearchCV(r, p, cv=5, n_jobs=-1, verbose=True)
rg.fit(ndf, df['Target'])
rg.best_params_["alpha"]
```

```python
clf = linear_model.Ridge(random_state=0, alpha = rg.best_params_["alpha"])
clf.fit(ndf, df['Target'])
pred = clf.predict(X_test)
print(mean_squared_error(pred, y_test))
pred = clf.predict(X_train)
print(mean_squared_error(pred, y_train))
pred = clf.predict(ndf)
print(mean_squared_error(pred, df['Target']))
```
